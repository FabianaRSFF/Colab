Análise de dados exploratória para conhecimento (empenho em diárias em municípios gaúchos e PIB do município .

```python
import pandas as pd
import seaborn as srn
import statistics as sts
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
base = pd.read_csv('dados.csv', sep=';')
base
```

```python
base.shape
```

```python
agrupado = base.groupby(['MUNICIPIO']).size()
agrupado
```

```python
agrupado = base.groupby(['PIB']).size()
agrupado
```

```python
agrupado = base.groupby(['VALOREMPENHO']).size()
agrupado
```

```python
plt.hist(base.iloc[:, 1], bins=6)
plt.title("MUNICIPIOS")
plt.ylabel("VALOREMPENHO")
plt.xlabel("PIB")
```

```python

plt.scatter(base.VALOREMPENHO, base.PIB, color ='blue', facecolors ='none', marker = '*')
plt.title("MUNICIPIO")
plt.xlabel("PIB")
plt.ylabel("VALOREMPENHO")
```

```python
plt.plot(base.VALOREMPENHO, base.PIB)
plt.title('MUNICIPIOS')
plt.xlabel('PIB')
plt.ylabel('VALOREMPENHO')
```

```python
plt.plot(base.MUNICIPIO, base.PIB)
plt.title('MUNICIPIOS')
plt.xlabel('PIB')
plt.ylabel('MUNICIPIO')
```

```python
figura = plt.figure()
eixo = figura.add_subplot(1, 1, 1, projection='3d')
eixo.scatter(base.CODIGO, base.PIB, base.VALOREMPENHO)
eixo.set_xlabel('CODIGO')
eixo.set_ylabel('PIB')
eixo.set_zlabel('VALOREMPENHO')
```

O PIB dos municípios gaúchos, de acordo com estes dados, que não apresentam datas específicas, é muito baixo. E o valor de empenho é diretamente proporcional ao tamanho da cavalgadura do autocrata genocida. #ForaGenocida
